```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Define the ecosystem grouping system
ECOSYSTEM_GROUPS = {
    'Terrestrial': [
        'Environmental;Terrestrial',
        'Mixed, Environmental;Terrestrial'
    ],
    'Engineered': [
        'Engineered;Modeled',
        'Engineered;Bioreactor',
        'Engineered;Built environment',
        'Engineered;Wastewater',
        'Engineered;Solid waste',
        'Engineered;WWTP',
        'Engineered;Artificial ecosystem',
        'Engineered;Food production',
        'Engineered;Lab enrichment',
        'Engineered;Bioremediation',
        'Engineered;Industrial production',
        'Engineered;Biotransformation',
        'Engineered',
        'Mixed, Engineered;Artificial ecosystem',
        'Mixed, Engineered;Lab enrichment',
        'Environmental, Engineered;Terrestrial, Solid waste',
        'Engineered;Wastewater',
        'Engineered;WWTP'
    ],
    'Plant-associated': [
        'Host-associated;Plants',
        'Host-associated;Algae'
    ],
    'Aquatic': [
        'Environmental;Aquatic'
    ],
    'Animal-associated': [
        'Host-associated;Mammals',
        'Host-associated;Arthropoda: Insects',
        'Host-associated;Birds',
        'Host-associated;Fish',
        'Host-associated;Annelida',
        'Host-associated;Porifera',
        'Host-associated;Cephalochordata',
        'Host-associated;Mollusca',
        'Host-associated;Arthropoda: Crustaceans',
        'Host-associated;Invertebrates',
        'Host-associated, Environmental;Arthropoda, Terrestrial'
    ],
    'Microbial-associated': [
        'Host-associated;Microbial',
    ],
    'Human-associated': [
        'Host-associated;Mammals: Human'
    ]
}

def categorize_ecosystem(source):
    """Strict categorization with zero overlap"""
    if source == "Host-associated;Mammals: Human":
        return "Human_associated"
    
    for group, patterns in ECOSYSTEM_GROUPS.items():
        if any(pattern == source for pattern in patterns):
            return group
    
    return "Uncategorized"  # Return "Uncategorized" if no match is found
```

```python

# Load the CSV file
df = pd.read_csv('Figure_5_data')

# Display the first few rows of the dataframe to ensure it loaded correctly
df.head()
```

```python
# Extract the first two parts of the 'ecosystem' column and join them back together
df['ecosystem_first_two'] = df['ecosystem'].str.split(';').str[:2].str.join(';')

# Apply the categorize_ecosystem function to the new column
df['ecosystem_group'] = df['ecosystem_first_two'].apply(categorize_ecosystem)

# Save the updated DataFrame back to the same CSV file
df.to_csv('Figure_5_data', index=False)

# Optionally, print the first few rows to confirm the update
print(df.head())
```
